```python
#for finding Positive score, negative score, total words, polarity,subjectivity
positive=[]
negative=[]
wordss=[]
polarity=[]
subjectivity=[]
for i in input['cleared']:
  p=0
  n=0
  s=0
  for j in i :
    s+=1
  
    if j in positive_words:
      p+=1
      
    elif j in negative_words:
      n+=1
  polarr= (p-n)/((p+n)+0.000001)
  subject=(p+n)/(s+0.000001)


  subjectivity.append(subject)
  polarity.append(polarr)
  positive.append(p)
  negative.append(n)
  wordss.append(s)
      
```

```python
#finding syllabel
def syllable_count(word):
    count = 0
    vowels = "aeiouy"
    if word[0] in vowels:
        count += 1
    for index in range(1, len(word)):
        if word[index] in vowels and word[index - 1] not in vowels:
            count += 1
    if word.endswith("e"):
        count -= 1
    if count == 0:
        count += 1
    return count
```

```python
syllbl_count= []
complex_word=[]
for i in mydata['cleared']:
  syllabel_count= []
  complex=0
  for j in i:
    if syllable_count(j)>2:
      complex+=1
    syllabel_count.append(syllable_count(j))
  complex_word.append(complex)
  sum_syl_count = sum(syllabel_count)
  syllbl_count.append(sum_syl_count)


```
